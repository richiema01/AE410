## Problem 1

To solve the inviscid Burgers' equation using the method of characteristics, we first consider the general form of the equation:
$$
\frac{\partial u}{\partial t} + \frac{\partial}{\partial x}\left( f(u) \right) = 0,
$$
where $f(u) = \frac{u^2}{2}$. The characteristic curves along which the solution is constant are defined by the differential equations:
$$
\frac{dt}{ds} = 1, \quad \frac{dx}{ds} = u, \quad \frac{du}{ds} = 0.
$$

These equations come from considering that along the characteristics, the total derivative of $u$ with respect to $s$ (where $s$ is the parameter along the characteristic curve) is zero:
$$
\frac{du}{ds} = \frac{\partial u}{\partial t} \frac{dt}{ds} + \frac{\partial u}{\partial x} \frac{dx}{ds} = 0.
$$

Substituting the inviscid Burgers' equation into the above, we get:
$$
\frac{\partial u}{\partial t} \frac{dt}{ds} + u \frac{\partial u}{\partial x} \frac{dx}{ds} = 0.
$$

Choosing $\frac{dt}{ds} = 1$ gives us a time parameterization along the characteristic, and we are left with:
$$
\frac{\partial u}{\partial t} + u \frac{\partial u}{\partial x} = 0.
$$

Since $\frac{du}{ds} = 0$, we have $u$ is constant along the characteristic curve, thus $\frac{dx}{ds} = u$. So the equations governing the characteristic curves become:
$$
\frac{dt}{ds} = 1, \quad \frac{dx}{ds} = u, \quad \frac{du}{ds} = 0.
$$

These can be solved with the initial conditions $u(x,0) = u_0(x)$ to give:

The equation $ \frac{du}{ds} = 0 $ implies that $ u $ is constant along a characteristic curve, which can be written as $ u(s) = C $, where $ C $ is a constant determined by the initial condition, so $ C = u_0(x_0) $.

The equation $ \frac{dt}{ds} = 1 $ indicates that as we move along the characteristic curve parameterized by $ s $, $t $ increases at the same rate as $ s $. This allows us to treat $ s $ and $ t $ interchangeably for the parameterization of the characteristic curve.

The equation $ \frac{dx}{ds} = u$ can now be integrated taking into account that $ u $ is constant along the characteristic curve. Integrating, we get:
$$
x(s) = u \cdot s + x_0,
$$
where $ x_0 $ is the initial position when $ s = 0$, and $ u $ is the speed at which the characteristic curve moves in the $ x $-direction, which is equal to $ u_0(x_0) $.

At any time $ t $, the characteristic curve that passes through the point $ (x, t) $ will have started from the point $ (x_0, 0) $, where $ x_0 $ is the initial position. Since $ s $ and $ t $ can be used interchangeably, we can rewrite the characteristic curve equation as:

$$
x = u_0(x_0) \cdot t + x_0.
$$

Rearranging for $ x_0 $, we get:
$$
x_0 = x - u_0(x_0) \cdot t.
$$

This is the position $ x_0 $ along the $ x $-axis at time $ t = 0 $ from which the characteristic started.

Finally, since $ u $ is constant along the characteristic, the value of $ u $ at any point $ (x, t) $ is the same as it was at $ (x_0, 0) $. Therefore, the solution to the PDE along the characteristics can be written implicitly as:
$$
u(x, t) = u_0(x_0) = u_0(x - tu(x, t)),
$$

where $ u_0 $ is the initial velocity profile, and$ x_0 $is the starting position of the characteristic that passes through the point $ (x, t) $ at time $ t $. This is the implicit solution to the inviscid Burgers' equation with the given initial condition.


## Problem 2

#### Derivation
The inviscid Burgers' equation is given by:
\begin{equation}
    \frac{\partial u}{\partial t} + \frac{\partial}{\partial x}\left( \frac{u^2}{2} \right) = 0
\end{equation}
with the initial condition for the Riemann problem:
\begin{equation}
    u(x, 0) = \begin{cases}
    u_L & \text{if } x \leq 0 \\
    u_R & \text{if } x > 0
    \end{cases}
\end{equation}

Using the method of characteristics, we derive different solutions based on the initial conditions $u_L$ and $u_R$. For $u_L > u_R$, the solution contains a shock, while for $u_L < u_R$, a rarefaction wave forms.

For $u_L > u_R$, the solution to the Riemann problem is:
\begin{equation}
    u(x,t) = \begin{cases}
    u_L & \text{if } x \leq St \\
    u_R & \text{if } x > St
    \end{cases}
\end{equation}
with the shock speed $S$ given by:
\begin{equation}
    S = \frac{f(u_R) - f(u_L)}{u_R - u_L} =  \frac{\frac{u_R^2}{2} - \frac{u_L^2}{2}}{u_R - u_L} = \frac{1}{2}(u_L + u_R)
\end{equation}
This represents a discontinuity propagating at speed $S$.

For $u_L < u_R$, the solution to the Riemann problem satisfying the entropy condition is:
\begin{equation}
    u(x,t) = \begin{cases}
    u_L & \text{if } x \leq u_L t \\
    \frac{x}{t} & \text{if } u_L t < x \leq u_R t \\
    u_R & \text{if } x > u_R t
    \end{cases}
\end{equation}
This describes a continuous change in $u$ from $u_L$ to $u_R$.

The Rankine-Hugoniot condition for the shock wave is given by:
\begin{equation}
    S = \frac{[[f(u)]]}{[[u]]}
\end{equation}
where $[[\phi(x,t)]]$ corresponds to the jump of $\phi(x,t)$ across the discontinuity.


#### Solutions
Given the initial conditions
\begin{equation}
    u(x, 0) = \begin{cases}
    u_L, & \text{if } x < 0, \\
    u_R, & \text{if } x > 0,
    \end{cases}
\end{equation}
and the similarity variable $ \xi = \frac{x}{t} $, the fundamental solutions are as follows:


For $ u_L > u_R $, there is a shock wave traveling with speed $ S $ where $ S = \frac{1}{2}(u_L + u_R) $. The solution is
    $
        u(x,t) = \begin{cases}
        u_L, & \text{if } \xi < S, \\
        u_R, & \text{if } \xi > S.
        \end{cases}
    $


For $ u_R > u_L $, the solution is a rarefaction wave given by
    $
        u(x,t) = \begin{cases}
        u_L, & \text{if } \xi \leq u_L, \\
        \xi, & \text{if } u_L < \xi \leq u_R, \\
        u_R, & \text{if } \xi > u_R.
        \end{cases}
    $


If $ u_L = u_R $, the solution is stationary, and $ u(x,t) = u_L = u_R $ everywhere.


If a left-travelling shock wave is physically relevant, it would propagate to the left with speed $ S $ negative.


Similarly, a right-travelling shock wave would propagate to the right with speed $ S $ positive.



```python
def riemann_solver(uL, uR, xi):
    # shock wave solution
    if uL > uR:
        S = 0.5 * (uL + uR)  # shock speed
        if xi < S:
            return uL
        else:
            return uR
    
    # rarefaction wave
    elif uR > uL:
        if xi < uL:
            return uL
        elif xi > uR:
            return uR
        else:
            return xi  #inside the rarefaction fan

    else:  # uL == uR
        return uL


test_cases = [
    {'uL': 1, 'uR': 0, 'xi': -0.5},  # Should return uL before the shock
    {'uL': 1, 'uR': 0, 'xi': 0.75},  # Should return uR after the shock
    {'uL': 0, 'uR': 1, 'xi': 0.5},   # Should return xi within the rarefaction wave
    {'uL': 0.5, 'uR': 0.5, 'xi': 0.5}, # Should return uL (stationary solution)
]


results = []
for case in test_cases:
    result = riemann_solver(case['uL'], case['uR'], case['xi'])
    results.append(result)

results

```

## Problem 3

```python
import numpy as np
def function_average(a, b, F, K):
    h = (b - a) / K  #width of each subinterval
    total = F(a) + F(b)  # Start with the first and last terms

    for k in range(1, K):
        total += 2 * F(a + k * h)

    average = (1 / (b - a)) * (h / 2) * total
    return average
```

```python
def cell_averages(a, b, F, N, K):
    cell_width = (b - a) / N  #width of each cell
    averages = np.zeros(N)

    # Compute the average for each cell
    for i in range(N):
        cell_a = a + i * cell_width
        cell_b = cell_a + cell_width
        averages[i] = function_average(cell_a, cell_b, F, K)

    return averages

def function_F(x):
    return np.sin(10 * x)

Nx = 10

averages_K1 = cell_averages(0, 1, function_F, Nx, 1)
averages_K2 = cell_averages(0, 1, function_F, Nx, 2)
averages_K3 = cell_averages(0, 1, function_F, Nx, 3)
averages_K5 = cell_averages(0, 1, function_F, Nx, 5)

def analytical_average(a, b):
    domain = (a, b, 10000)
    return 1/(b-a) * (1/10 * (np.cos(10*a) - np.cos(10*b)))
    #return np.average(function_F(domain))

analytical_averages = np.array([analytical_average(i/Nx, (i+1)/Nx) for i in range(Nx)])

errors_K1 = averages_K1 - analytical_averages
errors_K2 = averages_K2 - analytical_averages
errors_K3 = averages_K3 - analytical_averages
errors_K5 = averages_K5 - analytical_averages
eNorms = [np.linalg.norm(errors_K1),np.linalg.norm(errors_K2),np.linalg.norm(errors_K3),np.linalg.norm(errors_K5)]

print(eNorms)
```

```python
import matplotlib.pyplot as plt

#plot the results and compare
def plot_comparisons(cell_centers, numerical_averages, analytical_averages, errors, K_values):
    fig, ax = plt.subplots(1, 1, figsize=(12, 10))
    #cell averages
    ax.plot(cell_centers, analytical_averages, 'k--', label='Analytical Average')
    for k, averages in zip(K_values, numerical_averages):
        ax.plot(cell_centers, averages, label=f'Numerical Average (K={k})')
    ax.set_title('Comparison of Analytical and Numerical Cell-Averages of $F(x) = \sin(10x)$')
    ax.set_xlabel('Cell Center')
    ax.set_ylabel('Average Value')
    ax.legend()


cell_centers = [(i + 0.5)/Nx for i in range(Nx)]
numerical_averages = [averages_K1, averages_K2, averages_K3, averages_K5]
K_values = [1, 2, 3, 5]
all_errors = [errors_K1, errors_K2, errors_K3, errors_K5]

plot_comparisons(cell_centers, numerical_averages, analytical_averages, all_errors, K_values)


fig = plt.figure(figsize=(12,10), tight_layout=True)
ax1 = fig.add_subplot(221)
ax1.plot(K_values, eNorms, '.-', label = '2-Norm Error')
ax1.set_xlabel('K')
ax1.set_ylabel('Error')
ax1.set_yscale('log')
ax1.set_title('Log Error vs. Integration Points per Cell')
ax1.legend()
```

As we can see, increasing the number of integration points per cell generally increases the accuracy of the numerical integration compared to the exact analytical averages. As we increase the amount of integration points per cell, the error between the exact decrease exponentially.

## Problem 4

```python
def function_average(a, b, F, K):
    h = (b - a) / K
    total = F(a) + F(b)
    for k in range(1, K):
        total += 2 * F(a + k * h)
    average = (1 / (b - a)) * (h / 2) * total
    return average

def riemann_solver(uL, uR, xi):
    if uL > uR:
        S = 0.5 * (uL + uR)
        if xi < S:
            return uL
        else:
            return uR
    elif uR > uL:
        if xi < uL:
            return uL
        elif xi > uR:
            return uR
        else:
            return xi
    else:
        return uL

def u_initial(x):
    sigma = 3/40
    return 1 / 10 + np.exp(-((x - 1/4)**2) / (2 * sigma**2))

def f_true(u):
    return u**2 / 2

Nx = 500  #number of cells
x = np.linspace(0, 1, Nx + 1)  #cell boundaries
dx = 1 / Nx
u0 = np.zeros(Nx)  #initial u values in cells

#initial cell averages
for i in range(Nx):
    u0[i] = function_average(x[i], x[i+1], u_initial, 100)

dt = 0.5 * dx / np.max(u0)  # time step size, based on the CFL condition
t_end = 1.5
Nt = int(t_end / dt)

U = np.zeros((Nt+1, Nx))
U[0, :] = u0

for n in range(1, Nt+1):
    #Ghost cells for boundary conditions
    U_ghost = np.zeros(Nx + 2)
    U_ghost[1:-1] = U[n-1, :]
    #Left boundary
    U_ghost[0] = U_ghost[1]
    #Right boundary
    U_ghost[-1] = 2 * U_ghost[-2] - U_ghost[-3]
    
    #fluxes at cell boundaries
    F = np.zeros(Nx + 1)
    for i in range(1, Nx):
        F[i] = f_true(riemann_solver(U_ghost[i], U_ghost[i+1], 0))
    #boundary fluxes
    F[0] = f_true(riemann_solver(U_ghost[0], U_ghost[1], 0))
    F[-1] = f_true(riemann_solver(U_ghost[-2], U_ghost[-1], 0))
    
    #update cells
    for i in range(Nx):
        U[n, i] = U[n-1, i] - dt/dx * (F[i+1] - F[i])


times = np.arange(0,1.6, .1)
indices = [int(t / dt) for t in times]
cell_centers = x[:-1] + dx/2  #centers of cells for plotting

plt.figure(figsize=(12, 6))
j = 0
for i, index in enumerate(indices):
    plt.plot(cell_centers, U[index, :], label='t=' + str(j/10))
    j += 1

plt.title('Solution to Burgers Equation over Time')
plt.xlabel('x')
plt.ylabel('u')
plt.legend()
plt.grid(True)
plt.show()
```
